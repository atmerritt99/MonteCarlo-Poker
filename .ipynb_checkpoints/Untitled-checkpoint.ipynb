```python
import random

suits = {0:"C", 1:"H", 2:"S", 3:"D"}
faceCards = {10: "T", 11:"J", 12:"Q", 13:"K", 14:"A"}
faceCardValues = {"T": 10, "J": 11, "Q": 12, "K": 13, "A": 14}

def HighCard(hand):
    firstCard = hand[0]
    if firstCard[0].isdigit():
        maxValue = int(firstCard[0])
    else:
        maxValue = faceCardValues[firstCard[0]]
        
    for card in hand[1:]:
        if card[0].isdigit():
            value = int(card[0])
        else:
            value = faceCardValues[card[0]]
        if value > maxValue:
            maxValue = value
    return maxValue      

def Pair(hand):
    cardCount = {}
    for card in hand:
        value = card[0]
        if value in cardCount.keys():
            cardCount[value] = cardCount[value] + 1
        else:
            cardCount[value] = 1
    
    for count in cardCount:
        if cardCount[count] == 2:
            return True
    return False

def TwoPair(hand):
    cardCount = {}
    for card in hand:
        value = card[0]
        if value in cardCount.keys():
            cardCount[value] = cardCount[value] + 1
        else:
            cardCount[value] = 1
    
    onePair = False
    for count in cardCount:
        if cardCount[count] == 2:
            if onePair:
                return True
            else:
                onePair = True
    return False

def FullHouse(hand):
    cardCount = {}
    for card in hand:
        value = card[0]
        if value in cardCount.keys():
            cardCount[value] = cardCount[value] + 1
        else:
            cardCount[value] = 1
    
    onePair = False
    ThreeOfAKind = False
    for count in cardCount:
        if cardCount[count] == 2:
            if ThreeOfAKind:
                return True
            else:
                onePair = True
        elif cardCount[count] == 3:
            if onePair:
                return True
            else:
                ThreeOfAKind = True

    return False

def ThreeOfAKind(hand):
    cardCount = {}
    for card in hand:
        value = card[0]
        if value in cardCount.keys():
            cardCount[value] = cardCount[value] + 1
        else:
            cardCount[value] = 1
    
    for count in cardCount:
        if cardCount[count] == 3:
            return True
    return False

def FourOfAKind(hand):
    cardCount = {}
    for card in hand:
        value = card[0]
        if value in cardCount.keys():
            cardCount[value] = cardCount[value] + 1
        else:
            cardCount[value] = 1
    
    for count in cardCount:
        if cardCount[count] == 4:
            return True
    return False

def Straight(hand):
    values = []
    for card in hand:
        value = card[0]
        if value.isdigit():
            values.append(int(value))
        else:
            values.append(faceCardValues[value])
            
    values.sort()
    previousValue = values[0]
    
    for value in values[1:]:
        if value != previousValue + 1:
            return False
        else:
            previousValue = value
            
    return True

def Flush(hand):
    firstCard = hand[0]
    firstSuit = firstCard[1]
    for card in hand[1:]:
        suit = card[1]
        if firstSuit != suit:
            return False
    return True

def StraightFlush(hand):
    return Straight(hand) and Flush(hand)

def PokerHand(hand):
    if StraightFlush(hand):
        return 9
    elif FourOfAKind(hand):
        return 8
    elif FullHouse(hand):
        return 7
    elif Flush(hand):
        return 6
    elif Straight(hand):
        return 5
    elif ThreeOfAKind(hand):
        return 4
    elif TwoPair(hand):
        return 3
    elif Pair(hand):
        return 2
    else:
        return 1

cards = []

for i in range(4):
    suit = suits[i]
    for j in range(2,15):
        if j > 9:
            cards.append(faceCards[j] + suit)
        else:
            cards.append(str(j) + suit)

userInput = input("How many players are there (excluding you) >> ")
numPlayers = int(userInput)
yourHand = []

print("Enter each card in your hand individually.")

for i in range(5):
    yourHand.append(input("Enter a card >> "))
    cards.remove(yourHand[i])
    
yourValue = PokerHand(yourHand)

sims = 10000
draws = 0
losses = 0
for i in range(sims):
    random.shuffle(cards)
    start = 0
    end = 5
    equalValue = False
    loss = False
    for j in range(numPlayers):
        value = PokerHand(cards[start:end])
        if yourValue < value:
            losses = losses + 1
            loss = True
            break
        elif yourValue == value:
            equalValue = True
        start = start + 5
        end = end + 5
    if equalValue and not loss:
        draws = draws + 1
print("Chance of Winning: " + str((sims-losses-draws)/sims*100))
print("Chance of Draw: " + str(draws/sims*100))
print("Chance of Loss: " + str(losses/sims*100))
```

```python
import random

#Generating Card Deck
suits = {0:"C", 1:"H", 2:"S", 3:"D"}
faceCards = {10: "T", 11:"J", 12:"Q", 13:"K", 14:"A"}
faceCardValues = {"T": 10, "J": 11, "Q": 12, "K": 13, "A": 14}

cards = []

for i in range(4):
    suit = suits[i]
    for j in range(2,15):
        if j > 9:
            cards.append(faceCards[j] + suit)
        else:
            cards.append(str(j) + suit)

userInput = input("How many players are there (excluding you) >> ")
numPlayers = int(userInput)
yourHand = []
playersHands = []

random.shuffle(cards)

print("Enter each card in your hand individually.")

for i in range(5):
    yourHand.append(input("Enter a card >> "))
    
for i in range(numPlayers):
    playerHand = []
    for j in range(5):
        playerHand.append(cards.pop())
    playersHands.append(playerHand)
    
print(yourHand)
print(playersHands)
```
